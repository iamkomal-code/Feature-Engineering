```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv('train.csv',usecols=['X1','X2','X3','X4','X5','X6'])
```

```python
df.head()
```

```python
#how many labels each variable has
for col in df.columns:
    print(col,':' ,len(df[col].unique()),'labels')
```

```python
for col in df.columns:
    print(col,':' ,df[col].value_counts(),'labels')
```

```python
#lets examine the how many columns we will obtain after one hot encoding
pd.get_dummies(df,drop_first=True).shape
```

we can see that from 6 initial categorical variables,we end up with 117 new variables

```python
#lets find the top 10 frequent categories for the variable x2

df.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
#lets make a list with most frequent categories of the variable
top_10 = [x for x in df.X2.value_counts().sort_values(ascending=False).head(10).index]
```

```python
top_10
```

```python
#now we make the 10 binary variables

for label in top_10:
    df[label] = np.where(df['X2']==label,1,0)
df[['X2']+top_10].head()
```

```python
#get whole set of dummy variables,for all the categorical variables

def one_hot_top_x(data,variable,top_x_labels):
    for label in top_x_labels:
        data[variable+'_'+label]=np.where(df[variable]==label,1,0)
        
df = pd.read_csv('train.csv',sep=",", usecols=['X1','X2','X3','X4','X5','X6'])

#encode x2 into 10 most frequent categories
one_hot_top_x(df,'X2',top_10)
df.head()
```

```python
top_10 = [x for x in df.X1.value_counts().sort_values(ascending=False).head(10).index]
one_hot_top_x(df,'X1',top_10)
df.head()
```
